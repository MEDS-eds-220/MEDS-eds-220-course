```python
import os

import pandas as pd
import geopandas as gpd


# ------ IMPORT DATASETS ------
# WDPA loading takes ~5 minutes. 

data_folder = os.path.join('data', 'WDPA_Nov2025_Public_shp')

fp = os.path.join(data_folder, 'WDPA_Nov2025_Public_shp_0', 'WDPA_Nov2025_Public_shp-polygons.shp')
wdpa0 = gpd.read_file(fp)
```

```python

fp = os.path.join(data_folder, 'WDPA_Nov2025_Public_shp_1', 'WDPA_Nov2025_Public_shp-polygons.shp')
wdpa1 = gpd.read_file(fp)
```

```python

fp = os.path.join(data_folder, 'WDPA_Nov2025_Public_shp_2', 'WDPA_Nov2025_Public_shp-polygons.shp')
wdpa2 = gpd.read_file(fp)
```

```python
# ------ SUBSET AND SAVE LATIN AMERICA DATA ------
latin_america_iso3 = [
    # North & Central America
    "MEX", "BLZ", "GTM", "SLV", "HND", "NIC", "CRI", "PAN",
    # Caribbean
    "CUB", "HTI", "DOM", "JAM", "TTO", "BHS", "BRB", "DMA", "GRD", "LCA",
    "VCT", "ATG", "KNA",
    # South America
    "COL", "VEN", "GUY", "SUR", "ECU", "PER", "BOL", "BRA", "PRY", "URY", "ARG", "CHL"
]

for df, i in zip([wdpa0, wdpa1, wdpa2], [0,1,2]):

    df_latam = df[df['ISO3'].isin(latin_america_iso3)]

    fp = os.path.join('data', 
                      'WDPA_Nov2025_Latam',
                      f'WDPA_Nov2025_Latam_shp_{i}',
                      f'WDPA_Nov2025_Latam_shp{i}-polygons.shp')
    df_latam.to_file(fp)
    
    df_read = gpd.read_file(fp)             # Verify saved file can be read back correctly
    assert len(df_latam) == len(df_read)
```
